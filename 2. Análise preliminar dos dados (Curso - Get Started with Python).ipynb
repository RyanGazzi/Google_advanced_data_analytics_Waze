### **Task 2c. Valores nulos**

Crie um dataframe somente das linhas ausentes e compare estatisticamente com o restante a fim de descobrir possíveis gaps


```python
# Isolate rows with null values
null_df = df[df['label'].isnull()]

# Display summary stats of rows with null values
null_df.describe()
```

```python
# Isolate rows without null values
not_null_df = df[~df['label'].isnull()]

# Display summary stats of rows without null values
not_null_df.describe()
```

#### **Resultados:**
Os valores estatísticos são parecidos, não irá interferir de forma significativa na análise

### **Task 2d. Por dispositivo**

Cheque as populações de 'iphone' e 'android' no dataframe de linhas ausentes, calcule o percentual de cada, compare com o dataset inicial

```python
# Get count of null values by device
print('Iphone ' + str(null_df[null_df['device'] == 'iPhone'].count().values[0]))
print('Iphone ' + str(null_df[null_df['device'] != 'iPhone'].count().values[0]))
```

```python
# Calculate % of iPhone nulls and Android nulls
null_df['device'].value_counts(normalize = True)
```

```python
# Calculate % of iPhone users and Android users in full dataset
df['device'].value_counts(normalize = True)
```

#### **Resultados:**
A porcentagem de valores ausentes por cada dispositivo é consistente com sua representação nos dados em geral. Não há nada que sugira uma causa não aleatória dos dados ausentes.

### **Task 2d. Por label**
Calcule o percentual de retenção x abandono e também a média dos valores para cada um deles


```python
# Calculate counts of churned vs. retained
df['label'].value_counts(normalize = True)

```

```python
# Calculate median values of all columns for churned and retained users
df.groupby('label').median(numeric_only = True)
```

Calcule a mediana de quilômetros por viagem no último mês para usuários retidos e com rotatividade.

```python
# Add a column to df called `km_per_drive`
df['km_per_drive'] = df['driven_km_drives'] / df['drives']

# Group by `label`, calculate the median, and isolate for km per drive
df.groupby('label').median()
```

Calcule a media de quilômetros por dia dirigido

```python
# Add a column to df called `km_per_driving_day`
df['km_per_driving_day'] = df['driven_km_drives'] / df['driving_days']

# Group by `label`, calculate the median, and isolate for km per driving day
df.groupby('label').median()
```

Calcule o número médio de viagens por dia dirigido

```python
# Add a column to df called `drives_per_driving_day`
df['drives_per_driving_day'] = df['drives'] / df['driving_days']

# Group by `label`, calculate the median, and isolate for drives per driving day
df.groupby('label').median()
```

Calcule a rotatividade por tipo de dispositivo

```python
# For each label, calculate the number of Android users and iPhone users
df.groupby(['label', 'device']).size()
```

```python
# For each label, calculate the percentage of Android users and iPhone users
df.groupby('label')['device'].value_counts(normalize=True)
```

#### **Resultados:**
1. O percentual de usuários por dispositivo comparando o dataset nulo com o geral é bem semelhante. Não há nada que sugira que as linhas ausentes não sejam aleatórias.  
2. O dataset contém cerca de 82% de usuário retidos e 18% de abandonos.  
3. Sugiro tomar cuidado ao realizar cálculos de médias neste dataset, utilizei a mediana devido uma grande variação entre o mínimo e o máximo no describe, diria que quase impossível de ocorrer.  
4. Os usuários que abandonaram o aplicativo dirigiram cerca de 240% a mais que os retidos.  
5. Independente de retidos ou não, os usuários representados neste dataset dirigiram bastante.  
6. A proporção entre usuários de iphone e android é consistente entre os retidos e abandonos.
